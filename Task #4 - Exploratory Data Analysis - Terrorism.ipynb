```python
# Importing the necessary libraries
import pandas as pd

# Loading the dataset
df = pd.read_csv("C:/Users/LENOVO/Downloads/globalterrorismdb_0718dist.csv",encoding='ISO-8859-1')

# Displaying the first rows of the dataset to understand its structure
print("The first rows of the dataset: :")
print(df.head())

# Viewing information about the dataset to understand its characteristics
print("\nInformation about the dataset :")
df.info()
```

```python
df.head(10)
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
import matplotlib.pyplot as plt

# Group by the 'iyear' column and plot the number of attacks per year
df.groupby('iyear').size().plot(kind='line', title='Terrorist Attacks Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Plot the top 10 affected countries
df['country_txt'].value_counts().head(10).plot(kind='bar', title='Top 10 Affected Countries')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Attack types distribution
df['attacktype1_txt'].value_counts().plot(kind='bar', title='Attack Types Distribution')
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.show()

# Target types distribution
df['targtype1_txt'].value_counts().plot(kind='bar', title='Target Types Distribution')
plt.xlabel('Target Type')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Top 10 terrorist groups
df['gname'].value_counts().head(10).plot(kind='bar', title='Top 10 Terrorist Groups')
plt.xlabel('Group Name')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Aggregate number of attacks by region
attacks_by_region = df['region_txt'].value_counts()

# Aggregate number of attacks by country
attacks_by_country = df['country_txt'].value_counts()

print("Top 10 Regions by Number of Attacks:")
print(attacks_by_region.head(10))

print("\nTop 10 Countries by Number of Attacks:")
print(attacks_by_country.head(10))
```

```python
import seaborn as sns
# Plot the top 10 regions
plt.figure(figsize=(10, 6))
sns.barplot(x=attacks_by_region.head(10).index, y=attacks_by_region.head(10).values, palette='viridis')
plt.title('Top 10 Regions by Number of Terrorist Attacks')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=90)
plt.show()

# Plot the top 10 countries
plt.figure(figsize=(10, 6))
sns.barplot(x=attacks_by_country.head(10).index, y=attacks_by_country.head(10).values, palette='viridis')
plt.title('Top 10 Countries by Number of Terrorist Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define grid dimensions
lon_bins = np.linspace(-180, 180, 100)
lat_bins = np.linspace(-90, 90, 50)

# Compute 2D histogram of attack locations
hist, xedges, yedges = np.histogram2d(df['longitude'], df['latitude'], bins=[lon_bins, lat_bins])

# Plot heatmap
plt.figure(figsize=(12, 8))
plt.imshow(hist, interpolation='nearest', origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], cmap='Reds')
plt.colorbar(label='Number of Attacks')
plt.title('Heatmap of Terrorist Attacks')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(False)
plt.show()

```

-Terrorist Attacks Over the Years: The line plot shows a general increasing trend in terrorist attacks over the years, with some fluctuations.
-Top 10 Affected Countries: The bar chart reveals that countries like Iraq, Pakistan, and Afghanistan have experienced a high number of terrorist attacks.
