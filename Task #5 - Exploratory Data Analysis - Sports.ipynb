```python
import pandas as pd

# Load the CSV files into pandas merged_dfFrames
matches_df = pd.read_csv('C:/Users/LENOVO/Downloads/matches.csv')
deliveries_df = pd.read_csv('C:/Users/LENOVO/Downloads/deliveries.csv')
matches_df.head()

```

```python
deliveries_df.head()
```

```python
# Merge the merged_dfFrames based on the match ID
merged_df = pd.merge(matches_df, deliveries_df, left_on='id', right_on='match_id', how='inner')

# Display the merged merged_dfFrame
merged_df.head()
```

```python
merged_df.columns
```

```python
print(merged_df.shape)
# Check for duplicate rows
duplicate_rows = merged_df[merged_df.duplicated()]

# Print duplicate rows
print("Duplicate Rows:")
print(len(duplicate_rows))

# Remove duplicate rows
cleaned_merged_df = merged_df.drop_duplicates()

# Print the shape of the cleaned merged_dfset
print("Shape of cleaned merged_dfset:", cleaned_merged_df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Explore the distribution of key variables
# Runs scored
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['batsman_runs'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Runs Scored')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()

# Wickets taken
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['player_dismissed'].dropna(), bins=20, kde=False, color='green')
plt.title('Distribution of Wickets Taken')
plt.xlabel('Player Dismissed')
plt.ylabel('Frequency')
plt.show()

# Match outcomes
plt.figure(figsize=(8, 5))
merged_df['winner'].value_counts().plot(kind='bar', color='orange')
plt.title('Match Outcomes')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.show()

# Analyze performance of different teams and players over different seasons
plt.figure(figsize=(10, 6))
sns.countplot(x='season', data=merged_df, palette='viridis')
plt.title('Matches Played Over Different Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Matches Played')
plt.xticks(rotation=90)
plt.show()

# Investigate factors such as toss decisions and venue on match outcomes
# Toss decisions
plt.figure(figsize=(8, 5))
sns.countplot(x='toss_decision', data=merged_df, palette='Set2')
plt.title('Toss Decisions')
plt.xlabel('Decision')
plt.ylabel('Frequency')
plt.show()

# Venue
plt.figure(figsize=(12, 6))
venue_counts = merged_df['venue'].value_counts().head(10)
venue_counts.plot(kind='bar', color='salmon')
plt.title('Top 10 Venues with Most Matches')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()


# Identify trends and patterns in the data
# Season-wise runs scored
season_runs = merged_df.groupby('season')['batsman_runs'].sum()
plt.figure(figsize=(10, 6))
season_runs.plot(kind='line', marker='o', color='purple')
plt.title('Total Runs Scored Each Season')
plt.xlabel('Season')
plt.ylabel('Total Runs Scored')
plt.xticks(rotation=45)
plt.show()
```

```python
# Drop non-numeric columns for correlation matrix
numeric_data = merged_df.select_dtypes(include='number')

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Check for columns with little variance 
low_variance_cols = numeric_data.var()[numeric_data.var() < 1e-2].index
numeric_data = numeric_data.drop(columns=low_variance_cols)

# Plotting the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Statistical analysis
# Summary statistics for key variables
summary_stats = numeric_data.describe()
print("Summary Statistics:")
print(summary_stats)
```

```python
# Compute the correlation matrix
correlation_matrix = numeric_data.corr()

# Convert the correlation matrix to a DataFrame for better readability
correlation_table = pd.DataFrame(correlation_matrix)

# Display the correlation table
print("Correlation Table:")
print(correlation_table)
```

```python
# Number of wins by each team
team_wins = matches_df['winner'].value_counts()
print("Number of wins by each team:")
print(team_wins)

# Plotting the number of wins by each team
plt.figure(figsize=(12, 6))
team_wins.plot(kind='bar', color='blue')
plt.title('Number of Wins by Each Team')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.show()
```

Analyze Team Performance

```python
# Winning margins by runs
win_by_runs = matches_df[matches_df['win_by_runs'] > 0]

plt.figure(figsize=(12, 6))
sns.histplot(win_by_runs['win_by_runs'], bins=30, kde=True, color='red')
plt.title('Distribution of Winning Margins by Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()

# Winning margins by wickets
win_by_wickets = matches_df[matches_df['win_by_wickets'] > 0]

plt.figure(figsize=(12, 6))
sns.histplot(win_by_wickets['win_by_wickets'], bins=10, kde=True, color='green')
plt.title('Distribution of Winning Margins by Wickets')
plt.xlabel('Wickets')
plt.ylabel('Frequency')
plt.show()
```

Analyze Player Performance

```python
# Top Batsmen
top_batsmen = deliveries_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
print("Top 10 Batsmen by Runs Scored:")
print(top_batsmen)

# Plotting top batsmen
plt.figure(figsize=(12, 6))
top_batsmen.plot(kind='bar', color='purple')
plt.title('Top 10 Batsmen by Runs Scored')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()

# Top Bowlers
top_bowlers = deliveries_df[deliveries_df['player_dismissed'].notna()].groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10)
print("Top 10 Bowlers by Wickets Taken:")
print(top_bowlers)

# Plotting top bowlers
plt.figure(figsize=(12, 6))
top_bowlers.plot(kind='bar', color='orange')
plt.title('Top 10 Bowlers by Wickets Taken')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.show()
```

Analyze Factors Influencing Match Outcomes

```python
# Toss decision impact on match outcome
toss_impact = matches_df.groupby('toss_decision')['winner'].count()
print("Impact of Toss Decision on Match Outcome:")
print(toss_impact)

# Plotting the impact of toss decisions
plt.figure(figsize=(8, 5))
toss_impact.plot(kind='bar', color='cyan')
plt.title('Impact of Toss Decision on Match Outcome')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Wins')
plt.show()
```

```python
# Number of wins by each team over different seasons
wins_per_season = matches_df.groupby(['season', 'winner']).size().unstack().fillna(0)

plt.figure(figsize=(15, 8))
wins_per_season.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Number of Wins by Each Team Over Different Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
